```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,200)
```

```python
x
```

```python
y=np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
# plot and xlabel and ylabel and title and show
plt.plot(x,y)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
```

```python
x
```

```python
y
```

```python
#random genterate data in 50 and scatter plot and size and color and aplha intensity/opacity
colours=np.random.rand(50)
size=100*np.random.rand(50)
plt.scatter(x,y,c=colours,s=size,alpha=.5,marker="x")
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
```

```python
# vertical barplot
x=["a","b","c","d","e"]
y=np.random.rand(5)
plt.bar(x,y)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
```

```python
x
```

```python
y
```

```python
#horizontal barplot
plt.barh(x,y)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
```

```python
#control figure size and dot slash line  in declrear in plot and grid is a line
x=np.linspace(0,10,200)
y=np.sin(x)
plt.figure(figsize=(6,2))
plt.grid()
plt.plot(x,y,'--b')
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
```
