Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
print("A scatter plot is a diagram where each value in the data set is represented by a dot.")
```

```python
np.random.seed(3)
```

```python
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
plt.plot(x,y)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
print("subplot() function adds subplot to a current figure at the specified grid position.  It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. So to create multiple plots you will need several lines of code with the subplot() function. ")
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
print("A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories")
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
print("A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.")
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
plt.boxplot(box2)
plt.xlabel("this is my x axis data")
plt.ylabel("this is my y axis data")
plt.title("this is my title")
plt.show()
```
