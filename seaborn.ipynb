```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
tips.smoker.value_counts()
```

```python
tips["smoker"].value_counts()
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,hue=tips.smoker)
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue=tips.smoker)
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue=tips.smoker,style=tips.day)
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue=tips.smoker,style=tips.sex)
```

```python
sns.catplot(x="sex",y="tip",data=tips)
```

```python
sns.pairplot(tips)
```
